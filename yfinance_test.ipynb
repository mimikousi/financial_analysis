```python
# Timestamp オブジェクト
end_eps = pd.Timestamp(eps.index[0])
end = pd.Timestamp(hist.index[-1])
start = end - pd.DateOffset(years=1)

# 文字列形式で日付を取得
start_str = start.strftime('%Y-%m-%d')
end_str = end.strftime('%Y-%m-%d')
end_eps_str = end_eps.strftime('%Y-%m-%d')
```

```python
roe = financials.T["Net Income"] / balance_sheet.T["Stockholders Equity"]
roe
```

```python
per = hist["Close"][start_str:end_str] / eps[end_eps_str]
per.tail(10)
```

```python
# グラフのサイズを設定（任意）
plt.figure(figsize=(10, 6))

# 散布図を作成：x軸に 'PER（会予）', y軸に '売上総利益率'
plt.plot(hist["Close"][start_str:end_str], label=f'{num}')

# 軸ラベルを設定
plt.ylabel('株価', fontsize=16)

# メモリラベルのフォントサイズを大きくする
plt.tick_params(axis='both', which='major', labelsize=14)
plt.legend()
plt.savefig(f'./price/{num}_stock_price.png', dpi=150, facecolor='white')
```
